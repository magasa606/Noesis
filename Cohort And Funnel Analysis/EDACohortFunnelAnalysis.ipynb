COHORT CREATION FOR COHORT ANALYSIS

A cohort is a group of individuals who share certain common characteristics or experiences (such as birth year, disease, or exposure) and are followed over time to study changes or outcomes. It is often used in medical research (cohort studies), social sciences, and business analysis (cohort analysis) to understand generational behavior or trends. The term originates from ancient Roman military units and is also similar to the concept of "generations" or age groups.

5️⃣ Why are both MUST be present?

Because the cohort index = the time interval between these two events.

cohort_index = activity_month - cohort_month

In human language:

“How many months after the user joined did this event occur?”

FIRST ACTIVITY DATE PER USER (MUST HAVE) : 

1. Determine How Many Active Users Are Per First Event : 

This is the Foundation of Cohorts, Without This Cohorts Cannot Be Created

```python
first_activity = (df_cohort.groupby('user_id')['created_at']
                  .min().reset_index()
                  .rename(columns={'created_at':'first_activity_date'}))
```

2. Create Cohort Based On Month - Usually Cohort Based On Month, So Now Every User_Id Are Recorded Based On Month From Their First Activity Event.

```python
first_activity['cohort_month'] = (first_activity['first_activity_date'].
                                  dt.to_period('M'))
                                  
```

3. Combine These Data first_activity To df_cohort

Reason Is We Want To See Later How Many User Per Event After First Event To The Last Event, 
In Other Words, We Want To See How Many User Based On the process of developing events per event after the first event.

```python
df_cohort = df_cohort.merge(first_activity[['user_id', 'cohort_month']],
                            on='user_id', how='left')
```

4. Create Activity Month. 

We Already Created Cohort Month When User Started Active Per-First Event, Now We Create How Many User Active Per Month Event.

```python
df_cohort['activity_month'] = df_cohort['created_at'].dt.to_period('M')
```

5. Create For Calculating Cohort Index. 

The Logic Is Like These Event Happen In Which Month Since User Joined The Event. These Calculation Representing Mathematic Logic Between Year And Logic For Calculating the distance between active events per user and the user's activity towards a particular event. 

```python
df_cohort['cohort_index'] = ((df_cohort['created_at'].dt.year - df_cohort['cohort_month'].dt.year)
                             * 12  + (df_cohort['created_at'].dt.month - df_cohort['cohort_month'].dt.month))
```

6. Create Retention User Table

These Code For Calculating Unique User_Id Per Event And First Activity Event, So It Will Not Duplicate, Because User_Id Can Have Many Event But The Point Is Not Calculate That, But To Calculate Every User Who Come Back Over And Over, Or In Other Words Calculate Every Unique User_Id In Every Event Since First Activity Event For Retention Valuation. 



```python
cohort_table = (df_cohort.groupby(
    ['cohort_month','cohort_index'])['user_id'].
    nunique().reset_index()
)
```

```python
cohort_table.head()
```

7. Create Cohort Pivot Table

These Code For Make It Easier Later For Calculating Retention Rate.


```python
cohort_pivot = cohort_table.pivot(
    index='cohort_month',
    columns='cohort_index',
    values='user_id'
)
cohort_size = cohort_pivot.iloc[:,0] # TAKE ALL FIRST ROW AND COLUMN AT INDEX 0
```

8. Create Retention Rate Using Divide With Cohort Size Per Every Row In cohort_pivot. axis = 0 means the row movement is down per column, whereas if axis = 1 means the movement is sliding per row.

```python
retention_rate = cohort_pivot.divide(cohort_size, axis=0)
```

Here's the conceptual reason why funnels are SEPARATED : 

Reason 1 — The single responsibility dataset principle

In healthy BI:

- Transaction dataset → for operations & revenue

- Funnel dataset → for behavioral conversions

- One dataset = one main question

------------------------------------------------------------------

Reason 2 — Avoiding double counting in Tableau

Tableau:

- It's easy to be "fooled" by repeated data

- If the funnel is in transaction data:

- COUNT, SUM, even LOD can be wrong if not careful

- Mentors usually value clean and defensible funnels, not complex ones.

-----------------------------------------------------------------

Reason 3 — Funnels are summaries by design

Funnels are not exploratory tables, but:

- “Summaries of user journeys”

Philosophically:

- Funnels = processed results

- Transactions = raw materials

Mixing the two often:

- Obscures the story

- Creates ambiguous insights

```python
df_2['sale_price'] = (df_2['sale_price'].astype(str).str.
                      replace(',', '.', regex=False).astype(float))

df_2['cost'] = (df_2['cost'].astype(str).str.replace(',', '.', regex=False).
                astype(float))
```

1. First Step After EDA, We Need To Make Data Aggregation For Profit Analysis Later On.


REVENUE = SALE_PRICE * QUANTITY

TOTAL_COST = QUANTITY * COST

PROFIT  = REVENUE - TOTAL_COST

```python
df_2['revenue'] = df_2['sale_price'] * df_2['quantity']
df_2['total_cost'] = df_2['quantity'] * df_2['cost']
df_2['profit'] = df_2['revenue'] - df_2['total_cost']
```

2. Build a Funnel (User-Level) Which Means Total Orders Per User.

```python
user_funnel = (df_2.groupby('user_id')['order_id'].
               nunique().reset_index(name='total_orders'))
```

3. Build A Funnel Stage Based On Category Orders.

```python
def funnel_stage(x):
    if x == 1:
        return '1 Order'
    elif x == 2:
        return '2 Orders'
    else:
        return '3+ Orders'
    

user_funnel['funnel_stage'] = user_funnel['total_orders'].apply(funnel_stage)
```

4. Build Agreggation Profit Per User Level.

```python
user_profit = (df_2.groupby('user_id')['profit'].
               sum().reset_index(name='total_profit'))
```

5. Build Data Demographic Based On Demographic Data Existing In This Dataset.

```python
user_demographic = df_2[['user_id', 'user_name', 'age', 'gender']].drop_duplicates()
```

6. (OPTIONAL) BUILD FOR QUANTITY PER USER LEVEL FOR FURTHER ANALYSIS

```python
user_quantity = (df_2.groupby('user_id')['quantity'].
                 sum().reset_index(name='total_quantity'))
```

7. (OPTIONAL) BUILD FAVORITE PRODUCT AND CATEGORY PER USER LEVEL BUT NOT DUPLICATE BECAUSE AFTER WE GOT THE FAVORITE FROM THOSE ENTITYS WE DROP THE UNFAVORITE SO IT WILL STAY PER USER LEVEL BUT YOU NEED TO KNOW HOW TO USE IT SO IT WILL NOT MAKE FALSE INSIGHT.

FYI : 

The main function of size() is to return the total number of elements in a Pandas object (DataFrame or Series). Logic & Functionality: The logic is very simple: it counts every data cell present.

- For Series (Single Column): It returns the total number of rows.

- For DataFrame (Table): It returns the total number of cells (rows x columns).

MORE FYI :

Why use .size() in pandas dataframe with groupby because it can count the number of rows in a group that appear how many times, so it is the same as the .value_counts() function only with a wider scope (can be multi-column and more univariate) while in .value_counts() it is the same but is usually more aimed at short EDA and Statistics Because it is Usually For frequency analysis in one column or for global columns.

More FYI :

Drop duplicates can appear because after sort_values ​​it only displays per user level which category / product_name favorite then drops duplicates of the same user_id after its first appearance because sort_values ​​has been done earlier.

```python
user_category = (df_2.groupby(['user_id', 'category']).size().
                 reset_index(name='count'))

favorite_category = (user_category.sort_values(['user_id', 'count'], ascending=False).
                     drop_duplicates('user_id')[['user_id', 'category']].
                     rename(columns={'category':'favorite_category'}))
```

```python
user_product = (df_2.groupby(['user_id','product_name']).
                            size().reset_index(name='count'))

favorite_product = (user_product.sort_values(['user_id','count'], ascending=False).
                    drop_duplicates('user_id')[['user_id','product_name']].
                    rename(columns={'product_name':'favorite_product'}))
```
